## Work on X company data

```python
weight=pd.merge(x_order,x_sku,on='SKU')
weight
```

```python
weight['Total weight as per X (KG)']= weight['Order Qty']*weight['Weight (g)']*.001  # .001 multiply to convery gram into KG
```

```python
weight.head(2)
```

```python
final_weight=weight.groupby('ExternOrderNo')['Total weight as per X (KG)'].sum().reset_index()
final_weight
```

```python
# Seprate Weight into bins of .5kg
bins=[0,.5,1,1.5,2,2.5,3,3.5,4]
labels=[.5,1,1.5,2,2.5,3,3.5,4]

final_weight['Weight slab as per X (KG)']=pd.cut(final_weight['Total weight as per X (KG)'],bins=bins,labels=labels)
```

```python
final_weight.head(2)
```

```python
final_weight=final_weight.rename(columns={'ExternOrderNo':'Order ID'})
```

```python
final_weight
```

```python
final=pd.merge(final_weight,z_invoice,on='Order ID')
final
```

```python
final=final.rename(columns={'Charged Weight':'Total weight as per Courier Company (KG)'})
```

```python
final
```

```python
bins=[0,.5,1,1.5,2,2.5,3,3.5,4]
labels=[.5,1,1.5,2,2.5,3,3.5,4]

final['Weight slab charged by Courier Company (KG)']=pd.cut(final['Total weight as per Courier Company (KG)'],bins=bins,labels=labels)
```

```python
# Expected charges as per X company

z_rates.head(2)
```

```python
z_rates['fwd_a_fixed'][0]
```

```python
total_charge=[]
for i in range(len(final)):
    fwd_cat= 'fwd_'+final['Delivery Zone as per X'][i]
    fwd_fixed=z_rates[fwd_cat+'_fixed'][0]
    fwd_additional=z_rates[fwd_cat+'_additional'][0]
    
    rto_cat= 'rto_'+final['Delivery Zone as per X'][i]
    rto_fixed=z_rates[rto_cat+'_fixed'][0]
    rto_additional=z_rates[rto_cat+'_additional'][0]
    
    if final['Type of Shipment'][i]=='Forward charges':
        if final['Weight slab as per X (KG)'][i]<=0.5:
            total_charge.append(fwd_fixed)
        else:
            additional_weight=(final['Weight slab as per X (KG)'][i]-0.5)/0.5
            total_charge.append(fwd_fixed+ additional_weight * fwd_additional)
    
    if final['Type of Shipment'][i]=='Forward and RTO charges':
        if final['Weight slab as per X (KG)'][i]<=0.5:
            total_charge.append(fwd_fixed + rto_fixed)
        else:
            additional_weight=(final['Weight slab as per X (KG)'][i]-0.5)/0.5
            total_charge.append(fwd_fixed+ additional_weight * (fwd_additional + rto_additional))
    
final['Expected Charge as per X (Rs.)']= total_charge
        
    
    
    
```

```python
final=final.rename(columns={"AWB Code":'AWB Number','Zone':'Delivery Zone charged by Courier Company','Billing Amount (Rs.)':'Charges Billed by Courier Company (Rs.)'   })
```

## correctly charged

```python
correct=final[final['Difference Between Expected Charges and Billed Charges (Rs.)']==0.0]
correctly_charged=['Total orders where X has been correctly charged',len(correct), round(sum(correct['Charges Billed by Courier Company (Rs.)'])) ]

over=final[final['Difference Between Expected Charges and Billed Charges (Rs.)']<0.0]
over_charged=['Total Orders where X has been overcharged',len(over),round(np.absolute(sum(over['Difference Between Expected Charges and Billed Charges (Rs.)'])))]

under=final[final['Difference Between Expected Charges and Billed Charges (Rs.)']>0.0]
under_charged=['Total Orders where X has been undercharged',len(under),round(np.absolute(sum(under['Difference Between Expected Charges and Billed Charges (Rs.)'])))]
```

```python
summary=pd.DataFrame([correctly_charged,over_charged,under_charged], columns=['Details','Count','Amount(Rs)'])
summary
```

```python
# Save output to excel file

with pd.ExcelWriter('output.xlsx') as writer:
    summary.to_excel(writer, sheet_name='Summary')
    final.to_excel(writer, sheet_name='Calculation')
```
